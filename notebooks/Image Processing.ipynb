```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import skimage as ski
from skimage import io, filter, morphology
from skimage.feature import corner_harris, corner_subpix, corner_peaks
import operator
```

```python
data = pd.read_csv('../data/partial_train.csv').astype(np.float64)
X, y = data[[col for col in data.columns if not 'label' in col]], data['label']

imgs = []
idxs = map(
    operator.itemgetter(0),
    [ np.random.choice(y[y == n].index,1) for n in sorted(y.unique())]
)
for row in X.iloc[idxs].values:
    im = row.reshape(28,-1)
    imgs.append(im)

for im in imgs:
#     im = filter.gaussian_filter(im,sigma=.5)
    im = filter.sobel(im)
    coords = corner_peaks(corner_harris(im), min_distance=3)
    coords_subpix = corner_subpix(image, coords, window_size=5)
    
    fig, ax = plt.subplots()
    ax.imshow(im, interpolation='nearest', cmap=plt.cm.gray)
    ax.plot(coords[:, 1], coords[:, 0], '.b', markersize=15)
    ax.plot(coords_subpix[:, 1], coords_subpix[:, 0], '+r', markersize=15)
    plt.show()
   
```
